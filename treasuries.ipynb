```python
import os
import pandas as pd
```

```python
directory_name='treasury_yield_data_test'
directory = os.fsencode(directory_name)
cnt=0
result_df=None
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".csv") : 
        path_file=os.path.join(directory_name, filename)
        print(path_file)
        df = pd.read_csv(path_file)
        if result_df is None:
            result_df=df
        else:
            result_df=pd.concat([result_df,df])
            print("Adding Data")
    
        continue
    else:
        continue
print("Done")
print(result_df)
```
